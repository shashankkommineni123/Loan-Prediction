# Visualization 

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

#Matplot Basics

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
plt.plot(x, y, 'r') # 'r' is the color red
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
plt.show()
```

```python
plt.plot(x, x**2, label="x**2")
plt.plot(x, x**3, label="x**3")
plt.legend()
plt.show()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# Distribution Plot 

```python
sns.distplot(tips['total_bill'])
```

### Distribution plots are of crucial importance for exploratory data analysis. They help us detect outliers and skewness, or get an overview of the measures of central tendency (mean, median, and mode).

## pairplot

pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns). 

```python
sns.pairplot(tips)
```

```python
# here we compare between males and females for all the dataframe , also we changed the palette
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

# Bar plot
A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. 

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

# Count plot 
Show the counts of observations in each categorical bin using bars. A count plot can be thought of as a histogram across a categorical, instead of quantitative, variable.

```python
sns.countplot(x='sex',data=tips)
```

# Box Plot
a box plot or boxplot (also known as box and whisker plot) is a type of chart often used in explanatory data analysis. Box plots visually show the distribution of numerical data and skewness through displaying the data quartiles (or percentiles) and averages.

```python
sns.boxplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

# Implot 
Seaborn Implot is a method used to draw the plot of scattering on the facet grid, regression model, and the facet data if fitting across the facet grid. 

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

# Heat Map 
 Correlation heatmaps are a type of plot that visualize the strength of relationships between numerical variables. Correlation plots are used to understand which variables are related to each other and the strength of this relationship.

```python
# Matrix form for correlation data
tips.corr()
sns.heatmap(tips.corr(),annot=True)
```

#Pie Chart 
Pie chart is circle divided to slices. Each slice represents a numerical value and has slice size proportional to the value.

```python
import plotly.express as ex
ex.pie(tips,names='sex',title='Males & Females distribution',hole=0.20)
```
